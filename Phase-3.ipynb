## Advance operation with Business Examples:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dta structure: [restaurant_id, 2021,20222,2023,2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #Paradise Biryani
    [2, 120000, 140000, 160000, 190000], #Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("==== Zomato sales analysis ====")
print("\n Sales data shape", sales_data.shape)
print(" Sample data for first 3 restaurant:\n",sales_data[:3])#(:,1:)

#Total sales per year
#print("\n",np.sum(sales_data,axis=0))
#print("\n",np.sum(sales_data[:, 1:],axis=0 or axis =1))
yearly_total=np.sum(sales_data[:, 1:], axis=0)
print("Yearly total:\n", yearly_total)


# Minimum sales per restaurant/per year
min_sales = np.min(sales_data[:,1:],axis=1)
print("Minimum sales per restaurant:\n",min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:,1:],axis=0)
print("Maximum sales per year:\n",max_sales)

#Average sales per restaurant:
avg_sales = np.mean(sales_data[:,1:], axis=1)
print("Average sales:\n",avg_sales)

# cummulative sales
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sum per retaurant:\n" ,cumsum)
mean_cum = np.mean(cumsum, axis=0)
print("Mean cummulative sum:\n",mean_cum)

#plotting graph for this data 
plt.figure(figsize = (10, 6)) #figure size tuple data entery
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cummulative sales across all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()

```

## Vectors operations:

```python
#1d vector 
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("vector addition:", vector1 + vector2)
print("\n Multiplication vector:", vector1 * vector2)
print("\nDot Product:", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2)))
print("Angle between two:",angle)
```
