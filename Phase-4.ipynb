```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.random.rand(3,3)
array3 =  np.zeros((4,4))
```

```python
np.save('array1.npy',array1)
np.save('array2.npy',array2)
np.save('array3.npy',array3)

```

```python
loaded_array1 = np.load('array1.npy')
print("Loaded array1:\n",loaded_array1)



```

```python
from PIL import Image

# Step 1: Open the image file
img = Image.open("numpy-logo.png").convert('RGB')  # Make sure this file is in your folder

# Step 2: Convert the image to a NumPy array and normalize
img_array = np.array(img) / 255.0  # scale values from 0–255 to 0–1

# Step 3: Save the array as a .npy file
np.save("numpy-logo.npy", img_array)

try:
    logo=np.load('numpy-logo.npy')
    
    #display
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)

    dark_logo = 1-logo #invert the colour of original
    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy dark logo")
    plt.grid(False)
    
    
except FileNotFoundError:
    print("numpy logo file not found")
```
